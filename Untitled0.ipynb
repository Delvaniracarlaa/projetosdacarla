```python
#36 senha
senha_correta = "1234"
tentativas = 0

while tentativas < 3:
    senha = input("Digite a senha: ")

    if senha == senha_correta:
        print("Acesso liberado")
        break
    else:
        tentativas += 1
        print("Senha incorreta")

if tentativas == 3:
    print("Acesso bloqueado")
```

```python
#37 desconto
valor = float(input("Valor da compra: "))

if valor <= 100:
    desconto = 0
elif valor <= 500:
    desconto = 0.05
elif valor <= 1000:
    desconto = 0.10
else:
    desconto = 0.15

total = valor * (1 - desconto)

print(f"Desconto: {desconto*100:.0f}%")
print(f"Valor final: R$ {total:.2f}")
```
