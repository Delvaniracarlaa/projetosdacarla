```python
#atividade 06/05/2026
#criar uma classe com tres objetos
class Animal:
  def __init__(self, especie, nome, sexo):
    self.especie= especie
    self.nome= nome
    self.sexo= sexo
  def apresentar(self):

    print("especie:", self.especie)
    print("nome:", self.nome)
    print("sexo:", self.sexo)

animal1 = Animal("Cachorro", "Thor", "Macho")
animal2= Animal("Gato", "Sonic", "Macho")
animal3= Animal("Papagaio", "Juju", "Fêmea")

print('Dados do Animal 1')
animal1.apresentar()

print()

print('Dados do Animal 2')
animal2.apresentar()

print()

print('Dados do Animal 3')
animal3.apresentar()
```

```python
#criar uma classe e aplicar um loop para inserir dados
class GeneroMusical:
  def __init__(self, generomusical, banda, artista):

    self.generomusical = generomusical
    self.banda= banda
    self.artista= artista

  def descrever(self):

    print("generomusical:", self.generomusical)
    print("banda:", self.banda)
    print("artista:", self.artista)

lista_generosmusicais= []

for i in range(1, 4):
  print(f"Cadastro do {i} Gênero Musical")

  generomusical= input("digite o nome do gênero musical: ")
  banda= input("digite o nome da banda: ")
  artista= input("digite o nome do artista: ")
  print()

  novo_genero= GeneroMusical(generomusical, banda, artista)

  lista_generosmusicais.append(novo_genero)
print("\n=== GÊNEROS MUSICAIS CADASTRADOS ===")
for genero in lista_generosmusicais:
  genero.descrever()
  print()
```
